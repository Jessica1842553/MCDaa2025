```python
# Crear nueva columna: promedio de notas
df['avg_grade'] = (df['G1'] + df['G2'] + df['G3']) / 3
display(df.avg_grade)
```

```python
# Modificar columna: convertir 'yes'/'no' a 1/0 en 'internet'
df['internet'] = df['internet'].map({'yes': 1, 'no': 0})
display(df.internet)
```

```python
# Eliminar una columna - Guardian
df = df.drop(columns=['guardian'])
display(df)
```

```python
# Filtrar estudiantes que reprobaron y tienen internet
df_filtered = df[(df['avg_grade'] <= 14) & (df['internet'] == 1)]
display(df_filtered)
```

```python
# Subconjunto con estudiantes mujeres
df_female = df[df['sex'] == 'F']
display(df_female)
```

```python
# Grupo de datos - Promedio de notas por genero
grouped_by_gen = df.groupby('sex')['avg_grade'].mean()
display(grouped_by_gen)
```
